```python
# '|' - or dla warunkow
df[(df['przedmiot'] == 'Wstęp do informatyki') | (df['osoba'] == 'Faliszewski Piotr')]
```

```python
# groupby podobnie jak w SQL (nie wiem czy sie przyda)
df.groupby('sala').count()
```
